##**Problem Statement 1:Smart Parking Lot Management System**

Design a function to manage a smart parking lot.

**The system should:**

* Accept vehicle entry and exit logs


* Calculate total parked vehicles


* Identify peak parking usage


* Alert if parking exceeds capacity


```python
def smart_parking(vechiles_logs):
  capacity=50
  count=0
  # check for the logs and count the number of vehiles in the parking
  for i in vechiles_logs:
    if i=='IN':
      count+=1
    else:
      count-=1
    # check for the capacity and show the status
  if count>0 and count<=capacity:
    print(f"Currently parked vehicles: {count}\n Parking status: Available")
  else:
    print(f"Currently parked vehicles: {count} \n Parking status: Unavailable ")

vehicle_logs=["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking(vehicle_logs)
```

##**Problem Statement 2:Online Food Delivery Time Estimator**

Create a function that estimates delivery time based on:
* Distance (km)


* Weather condition


* Traffic level


```python
def delivery_time_estimator(distance_km, traffic_level, weather):
    base_time = distance_km * 5
    traffic_delays = {"Low": 0,"Medium": 10,"High": 20}
    weather_delays = {"Clear": 0,"Rainy": 15,"Stormy": 30}
    # Calculate total delivery time
    if traffic_level in traffic_delays:
      traffic_delay = traffic_delays[traffic_level]
    else:
      traffic_delay = 0
    if weather in weather_delays:
      weather_delay = weather_delays[weather]
    else:
      weather_delay = 0
    total_time = base_time + traffic_delay + weather_delay
    # Display result
    print(f"Estimated Delivery Time: {int(total_time)} minutes")
distance = 8
traffic = "High"
weather = "Rainy"
delivery_time_estimator(distance, traffic, weather)
```

##**Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

Build a function that analyzes seat booking data and:
* Calculates occupancy percentage


* Determines if show is Housefull


* Suggests opening additional shows


```python
def theatre_occupancy_analyzer(total_seats, booked_seats):
    num_booked = sum(booked_seats)
    occupancy_percent = (num_booked / total_seats) * 100
    # check show status
    if occupancy_percent == 100:
        status = "Housefull"
    elif occupancy_percent >= 75:
        status = "Almost Full"
    elif occupancy_percent >= 50:
        status = "Moderate Occupancy"
    else:
        status = "Low Occupancy"
    print(f"Occupancy: {occupancy_percent:.0f}%")
    print(f"Show Status: {status}")
total_seats = 200
booked_seats = [1] * 150 + [0] * 50
theatre_occupancy_analyzer(total_seats, booked_seats)
```

##**Problem Statement 4:Cloud Server Load Classification System**
Create a function to classify server load based on CPU usage readings.

**Rules:**

* Average CPU < 50% → Normal


* 50%–80% → Warning


* 80% → Critical


```python
def server_load_classifier(cpu_readings):
    if len(cpu_readings) == 0:
        print("No CPU readings available")
        return
    avg_cpu = sum(cpu_readings) / len(cpu_readings)
    # Classify server status
    if avg_cpu < 50:
        status = "Normal"
        recommendation = "System operating within optimal parameters"
    elif avg_cpu <= 80:
        status = "Warning"
        recommendation = "Monitor system closely - consider scaling resources"
    else:
        status = "Critical"
        recommendation = "URGENT: Scale up resources immediately!"
    print(f"Average CPU Load: {avg_cpu:.0f}%")
    print(f"Server Status: {status}")
cpu_readings = [45, 60, 70, 85, 90]
server_load_classifier(cpu_readings)
```

##**Problem Statement 5:Smart Classroom Resource Usage Monitor**
Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.



```python
def classroom_resource_monitor(resource_usage):
    thresholds = {"Projector": 6, "AC": 8, "Lights": 10}
    overused_resources = []
    for resource, hours in resource_usage.items():
        if resource in thresholds:
            threshold = thresholds[resource]
        else:
            threshold = 8
        if hours > threshold:
            overused_resources.append(resource)
    if len(overused_resources) > 0:
        print("Overused Resources:", overused_resources[0])
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
resource_usage = {"Projector": 6, "AC": 9, "Lights": 4}
classroom_resource_monitor(resource_usage)
```

##**Problem Statement 6: Online Event Registration Capacity Controller**
Create a function that manages event registrations by:
* Tracking registrations


* Preventing overbooking


* Triggering waitlist mode


```python
def event_registration_controller(event_capacity, registrations):
    if registrations <= event_capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = event_capacity
        waitlisted = registrations - event_capacity
        status = "Closed"
    print(f"\nConfirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")
event_registration_controller(100, 105)
```
