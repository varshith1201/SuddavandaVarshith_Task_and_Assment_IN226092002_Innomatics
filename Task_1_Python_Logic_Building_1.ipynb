## 1: User Login Check
**Problem Statement:** Given a username and password, check whether login is successful.


```python
# Define correct credentials
username = "admin"
password = "1234"

# Get user input
input_username = input("Enter username: ")
input_password = input("Enter password: ")

print()

# Check if both username and password match
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

## 2: Pass / Fail Analyzer
**Problem Statement:** Analyze student results.

```python
# Student marks list
marks = [45, 78, 90, 33, 60]

# Initialize counters for pass and fail students
pass_count = 0
fail_count = 0

# Analyze each student's marks
print("student results...")

for i, mark in enumerate(marks, 1):
    if mark >= 50:
        pass_count += 1
        status = "PASS"
    else:
        fail_count += 1
        status = "FAIL"
    print(f"Student {i}: {mark} marks - {status}")

# Print the final result

print(f"\nTotal Students: {len(marks)}")
print(f"Total Pass: {pass_count}")
print(f"Total Fail: {fail_count}")
print(f"Pass Percentage: {(pass_count / len(marks)) * 100:.2f}%")
```

## 3: Simple Data Cleaner
**Problem Statement:** Clean and standardize user names.

```python
# Original names list with inconsistent formatting
names = [" Alice ", "bob", " CHARLIE "]

print("Original names list:")
print(names)
print()

cleaned_names = []

for name in names:
    # Remove leading and trailing spaces, then convert to lowercase
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Print the cleaned list
print("Cleaned names list:")
print(cleaned_names)
```

## 4: Message Length Analyzer
**Problem Statement:** Analyze message sizes.

```python
# List of messages to analyze
messages = ["Hi", "Welcome to the platform", "OK"]

# Maximum allowed message length before flagging
MAX_LENGTH = 10

print("Message Length Analysis\n")

# Analyze each message
for i, message in enumerate(messages, 1):
    message_length = len(message)
    
    # Check if message exceeds the maximum length
    if message_length > MAX_LENGTH:
        flag = " [Message length exceeds 10 characters]"
    else:
        flag = ""
    
    print(f"Message {i}: \"{message}\"")
    print(f"Length: {message_length} characters{flag}")
    print()
```

## 5: Error Message Detector
**Problem Statement:** Detect error messages from system logs.

```python
# System logs list
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

print("System Log Analysis")

# Display all logs
print("Logs:")
for i, log in enumerate(logs, 1):
    print(f"  {i}. {log}")

print()

# Count the number of "ERROR" entries
error_count = logs.count("ERROR")

# Print the total error count
print(f"Total Logs: {len(logs)}")
print(f"Total ERROR entries: {error_count}")

# Additional analysis for context
if error_count > 0:
    print(f"\nWarning: {error_count} errors detected in the system logs!")
else:
    print("\nNo errors detected in the system logs.")
```
