##**Problem 1: Smart Light Controller**

**Problem Statement**

Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
class Light:
  def __init__(self,light_name,action):
    self.light_name = light_name
    self.action = action
    # display the status of the light
  def status(self):
    print(f"The {self.light_name} Light is {self.action}")
light=Light("Bedroom","ON")
light.status()
```

##**Problem 2: Employee ID Card System**

**Problem Statement**

Create a class that stores employee name, ID, and department and displays ID card details.


```python
class Employee:
  def __init__(self,Ename,EID,Edept):
    self.Ename = Ename
    self.EID = EID
    self.Edept = Edept
  # display the id card details
  def display_id_card(self):
    print(f"Employee ID card\nName: {self.Ename}\nID: {self.EID}\nDepartment: {self.Edept}")
employee=Employee("Rahul","EMP102","IT")
employee.display_id_card()
```

##**Problem 3: Mobile Contact Record**

**Problem Statement**

Create a class to store contact name and phone number and display contact information.


```python
class Contact:
  def __init__(self,name,phone_number):
    self.name = name
    self.phone_number = phone_number
  # show the contact is saved details of that contact
  def display_contact(self):
    print(f"Contact saved\nName: {self.name}\nPhone: {self.phone_number}")
contact=Contact("Anita","9876543210")
contact.display_contact()
```

##**Problem 4: Product Price Tag Generator**

**Problem Statement**

Create a class that stores product name and price and prints a formatted price tag.


```python
class Product:
  def __init__(self,name,price):
    self.name = name
    self.price = price
  # display the product and price in rupees
  def product_display(self):
    print(f"Name: {self.name}\nPrice: ₹{self.price}")
product=Product("Headphones",2499)
product.product_display()
```

##**Problem 5: Movie Rating Display System**

**Problem Statement**

Create a class that stores movie name and rating and displays movie details.


```python
class Movie:
  def __init__(self,name,rating):
    self.name = name
    self.rating = rating
    # display the movie name and rating
  def display_movie(self):
    print(f"Movie: {self.name}\nRating: {self.rating}/5")
movie=Movie("Inception",4.8)
movie.display_movie()
```
