##**Problem Statement 1: Employee Performance Bonus Eligibility**

**Description:**

A company evaluates employee performance scores at the end of the year. You are given a dictionary containing employee names and their performance scores.



```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
emp_elegible_bonus=[]
# checks for the highest score
high_performance=max(employees.values())
# stores the elegible employee names in a list
for key,value in employees.items():
  if value>=high_performance:
    emp_elegible_bonus.append(key)
print("Top performers elegible for bonus:",",".join(emp_elegible_bonus),"(score:",high_performance,")")
```

##**Problem Statement 2: Search Query Keyword Analysis**

**Description:**

An e-commerce website stores customer search queries. You are given a search query sentence entered by a user.


```python
query=input("enter your search query:")
query=query.lower()
# removes the puntuations from the query
for char in query:
  query=query.replace(" ","") and query.replace(".","")
words=query.split()
keyword_frequency={}
# add frequencyto each word in query and stores it in the key word frequency dictionary
for index,word in enumerate(words):
  if word not in keyword_frequency:
    keyword_frequency[word]=1
  else:
    keyword_frequency[word]+=1
high_keyword_frequency={}
# filters the keyword whose frequence is greater than or equal to 2 and stores it in high frequency key words dictionary
for key,value in keyword_frequency.items():
  if value>=2:
    high_keyword_frequency[key]=value
print(high_keyword_frequency)
```

##**Problem Statement 3: Sensor Data Validation**

**Description:**

A factory collects sensor readings every hour. Each reading is stored in a list where the index represents the hour and the value represents the sensor reading.

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings=[]
for index,value in enumerate(sensor_readings):
  # check for the even readings and store it ad index, value pair in valid readings list
  if value % 2==0:
    valid_readings.append((index,value))
# print the valid readings with hour
print("Valid sensor readings (Hour, Value):",valid_readings)
```

##**Problem Statement 4: Email Domain Usage Analysis**

**Description:**

A company wants to analyze which email providers its users are using. You are given a list of employee email IDs.

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
email_provider={}
# extracts providers and stores its frequency
for i in emails:
    provider=i.split("@")[1]

    if provider not in email_provider:
        email_provider[provider]=1
    else:
        email_provider[provider]+=1
no_of_users=len(emails)
# calculate the persentage of each provider and print it
for provider,count in email_provider.items():
    percentage=(count/no_of_users)*100
    print(f"{provider}: {percentage:.0f}%")
```

##**Problem Statement 5: Sales Spike Detection**

**Description:**

A retail company tracks daily sales. Sudden spikes in sales may indicate promotions or unusual activity.


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
sum=0
# calculate daily average
for i in sales:
  sum=sum+i;
# calculate the sum
avg=sum/len(sales)
spike_sales={}
# check for the sakes which are more than 30% of the average and store them with day
for index,value in enumerate(sales):
  if value >= (avg+(avg*0.3)):
    spike_sales[index+1]=value
# print the spicked sale day and sold amount
for day, sale in spike_sales.items():
    print("Day", day, ":", sale)
```

##**Problem Statement 6: Duplicate User ID Detection**

**Description:**

A system stores user IDs during registration. Duplicate IDs can cause data integrity issues.

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
dublecate_ids={}
# check how many times each id is repeated
for i in user_ids:
  if i not in dublecate_ids:
    dublecate_ids[i]=1
  else:
    dublecate_ids[i]+=1
# print only the dublecate and repeated ids and how many time sits repeated
for key,value in dublecate_ids.items():
  if value>1:
    print(f"{key} -> {value}times")
```
