##**Problem 1: Social Media – Post Engagement Analyzer**

**Problem Statement**

Create a function to analyze engagement on a social media post.

**Rules**

* Calculate total likes
* If Total Likes ≥ 1000 → Viral Post
* Otherwise → Normal Engagement


```python
def post_engagement(likes):
    # Calculate total likes
    total_likes = 0
    for like in likes:
        total_likes += like

    print("Total Likes:", total_likes)

    # Check if the post is viral
    if total_likes >= 1000:
        print("Viral Post")
    else:
        print("Normal Engagement")

post_likes = [200, 300, 150, 400]
post_engagement(post_likes)
```

##**Problem 3: Agriculture – Rainfall Adequacy Checker**

**Problem Statement**

Create a function that analyzes seasonal rainfall data.

**Rules**
* Calculate average rainfall
* Average ≥ required level → Adequate Rainfall
* Otherwise → Inadequate Rainfall


```python
def rainfall_adequacy_checker(rainfall,required_level):
  # calculate the average rainfall
  avg_rainfall = sum(rainfall) / len(rainfall)
  print("Average Rainfall:", avg_rainfall)
  # check wheather the average rainfall is upto the required rainfall or not
  if avg_rainfall >= required_level:
    return ("Adequate Rainfall")
  else:
    return ("Inadequate Rainfall")

rainfall=[71,78,73,70,65,75]
required_level=60
result=rainfall_adequacy_checker(rainfall,required_level)
print(f"Rainfall status: {result}")
```

##**Problem 4: Social Media – Duplicate Account Detection**

**Problem Statement**

Create a function to detect duplicate usernames in a system.

**Rules**
* If duplicates exist → Flag accounts
* Else → Accounts are unique


```python
def duplicate_acc_check(accounts):
  # check for the duplicate usernames using set
  unique_accounts = set(accounts)
# returns yes if duplicate user names exists
  if len(accounts) != len(unique_accounts):
    return "yes"
  else:
    return "no"

accounts = ["asus1","intel","asus1","nvdia","intel"]
result=duplicate_acc_check(accounts)
print(f"Duplicate Accounts Found: {result}")
```

##**Problem 6: Agriculture – Premium Crop Price Filter**

**Problem Statement**

Create a function that filters crop prices above a premium threshold.

**Rules**
* Price > ₹2000 → Premium Crop
* Else → Regular Crop


```python
def crop_filter(crops):
  premium_crops = []
  # check for the premium crops form the given crops
  for crop in crops:
    if crop > 2000:
      premium_crops.append(crop)
  return premium_crops

crops=[2500,3200,1800,1200,1500]
premium_crops=crop_filter(crops)
print(f"Premium crops: {premium_crops}")
```

##**Problem 9: E-Learning – Student Attendance Eligibility System**

**Problem Statement**

Create a function to calculate attendance percentage and check exam eligibility.

**Rules**
* Attendance ≥ 75% → Eligible
* Attendance < 75% → Not Eligible


```python
def exam_eligible_status(attendance):
  count=0
  # count the number of days present
  for i in attendance:
    if i.lower()=="p":
      count+=1
  # calculate the attendance persentage
  attendance_percentage=(count/len(attendance))*100
  print(f"Attendance percentage: {attendance_percentage}")
  # check for the elegibility
  if attendance_percentage>=75:
      return "Eligible"
  else:
      return "Not eligible"

attendence=["p","a","p","p","p","a","p","p","p","p"]
result=exam_eligible_status(attendence)
print(f"Exam status: {result}")
```

##**Problem 10: Smart Electricity Bill Analyzer**

**Problem Statement**

You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.

**Billing Rules**
* First 100 units → ₹3 per unit
* Next 100 units (101–200) → ₹5 per unit
* Above 200 units → ₹7 per unit

**Usage Classification**
* Total bill < ₹500 → Low Usage
* ₹500 – ₹1500 → Moderate Usage
* ₹1500 → High Usage


```python
def calculate_electricity_bill(units):
    bill = 0
    # First 100 units
    if units <= 100:
        bill = units * 3
    # 101–200 units
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    # Above 200 units
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    # Usage Classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    return bill, usage_status


units_consumed = 250
bill_amount, status = calculate_electricity_bill(units_consumed)
print("Total Bill: ₹", bill_amount)
print("Usage Level:", status)
```
