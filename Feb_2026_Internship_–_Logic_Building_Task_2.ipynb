## **Problem 1: Unique Words in a Sentence**

**Problem** **Statement**

Given a sentence, identify all unique words and display:

  The count of unique words

  The unique words themselves

```python
sentence_input = input("Enter a sentence: ")
# Split into words and filter out empty strings
words = [word for word in sentence_input.split() if word]
# Get unique words using a set
unique_words =[]
for word in words:
  if word not in unique_words:
    unique_words.append(word)
# Count unique words
count = len(unique_words)
print(f"\nNumber of unique words: {count}")
print(f"Unique words: {unique_words}")
```

#**Problem 2: Highest Salary from Employee Data**

**Problem Statement**

Store employee names and their salaries, and determine which employee has the highest salary.


```python
employe_details={"bob":35000,"ally":30000,"sam":25000}
highest_salary=0
highest_salary_holder=""
for name,salary in employe_details.items():
  #check for the highest salary form the employee details
  if salary>highest_salary:
    highest_salary=salary
    highest_salary_holder=name

print("highest salary is",highest_salary_holder,"-",highest_salary)
```

#**Problem 3: Find Maximum and Minimum Values**

**Problem Statement**

Find the maximum and minimum values from a list of numbers.


```python
numbers=[50,30,21,60,49,80,100,56,25]
max=numbers[0]
min=numbers[0]
for num in numbers:
  #check for the maximum number and minimum number form the list
  if num>max:
    max=num
  elif num<min:
    min=num
print("list:",numbers)
print("max:",max)
print("min:",min)
```

#**Problem 5: Calculate Attendance Percentage**

**Problem Statement**

Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
attdance=["P","A","P","p","A","p","p"]
count=0
for i in attdance:
  #count the number of P's
  if i=="P" or i=="p":
    count+=1
attdence_persentage=count/len(attdance)*100
print("attendance percentage:",attdence_persentage)
```

#**Problem 6: Remove Duplicate Phone Numbers**

**Problem Statement**

You are given a list of phone numbers that may contain duplicates. Write a Python program to remove duplicate phone numbers and display only the unique phone numbers.


```python
phone_numbers = [9876543210, 9123456789, 9876543210,3698624014]
unique_numbers=[]
for num in phone_numbers:
  #check for the unique number and add it to list
  if num not in unique_numbers:
    unique_numbers.append(num)
print("unique numbers",unique_numbers)
```

#**Problem 7: Count Character Frequency**

**Problem Statement**

Count how many times each character appears in a given string.


```python
string=input("Enter the string:-")
character={}
for ch in string:
  #check for each character count in string and add it to the dictonary
  if ch in character:
    character[ch]+=1
  else:
    character[ch]=1
print(character)
```

#**Problem 10: Calculate Average Marks**

**Problem Statement**

Calculate the average marks of a student from a list of marks.


```python
marks_list=[100,35,80,95,40,67,35]
sum=0
#calculate the sum of all the marks
for marks in marks_list:
  sum=sum+marks
avg=sum/len(marks_list)
#round the average to 2 numbers after the desimal
round_avg=round(avg,2)
print("average is:-",round_avg)
```
