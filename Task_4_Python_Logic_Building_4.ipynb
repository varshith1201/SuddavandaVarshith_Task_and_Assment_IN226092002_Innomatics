##**Problem 1: Mobile Recharge Validation System**

**Problem Statement**

Create a function to validate a mobile recharge request.

**Rules:**
* Recharge amount must be ≥ ₹50


* Valid plans: 199, 299, 399, 599


* User should be allowed to retry if invalid



```python
def mobile_recharge():
    while True:
      # user input
      charge_amount=int(input("enter the recharge amount(199, 299, 399, 599):"))
      # if the condition in if becomes true print the message and continue i.e restart
      if charge_amount<50:
        print("recharge failed because minimum recharge should be Rs:50.00,try again")
        continue
      if charge_amount not in [199, 299, 399, 599]:
        print("recharge failed invalid plan,try again")
        continue
      return (f"rechatge successful of Rs:{charge_amount}")
      break


result=mobile_recharge()
print(result)
```

##**Problem 2: Inventory Reorder Alert System**

**Problem Statement**

Given a dictionary of products and stock quantities:

* Print Reorder Alert if stock < 15

* Print Stock OK otherwise

```python
def stock_status(stock):
  # checks status of the each product in disctionary and prints the status
  for product,quantity in stock.items():
    if quantity<15:
      print(f"Reorder {product}, current stock is- {quantity}")
    else:
      print(f"Stock OK for {product}")
# stock dictionary
stock={
    "Milk":15,
    "Honey":7,
    "Soaps":35,
    "Shampoo":3
}
stock_status(stock)
```

##**Problem 3: Student Result Processing System**

**Problem Statement**

Create a function that:

* Accepts a list of student marks


* Calculates average marks


* Prints Pass if average ≥ 50 else Fail

```python
def student_result(marks):
  sum=0
  for i in marks:
    sum=sum+i
  avg=sum/len(marks)
  if avg>=50 :
    return "pass"
  else:
    return "fail"
marks=[45,40,55,69,39,60]
result=student_result(marks)
print(result)
```

##**Problem 4: Cab Fare Estimator with Retry Option**

**Problem Statement**

Create a cab fare calculator:
* Base fare = ₹50


* ₹12 per km


* Peak hour → 25% extra


* Allow retry using while loop


```python
def cab_fair_cal():
  base_fee=50
  while True:
    # take the distance from the user as input
    distance=float(input("enter the distance in km:"))
    if distance<0:
      print("invalid distance,try again")
      continue
    # check for the peak time
    peak=input("is it peak hour(yes/no):").lower()
    # calculate the fare
    fee=base_fee+(distance*12)
    if peak=="yes":
      fee=fee+(fee*0.25)
    # conform the details from the user if not retry
    retry=input("conform the distance and peak time (yes/no):").lower()
    if retry=="yes":
      return fee
      break
result=cab_fair_cal()
print(f"the estimated fare for the cab is: {result}")
```

##**Problem 5: Employee Attendance Eligibility Checker**

**Problem Statement**

Create a function that:

* Accepts attendance list ("P", "A")


* Calculates attendance percentage


* Returns Eligible if ≥ 75% else Not Eligible

```python
def employee_attendance(attendance):
  count=0
  # count the number of days present
  for i in attendance:
    if i.lower()=="p":
      count+=1
  # calculate the attendance persentage
  attendance_percentage=(count/len(attendance))*100
  # check for the elegibility
  if attendance_percentage>=75:
      return "eligible"
  else:
      return "not eligible"
# user input
attendance=["p","a","p","p","p"]
result=employee_attendance(attendance)
print(result)
```

##**Problem 6: Password Strength Checker**

**Problem Statement**

Create a function to check password strength

**Rules:**
* Minimum length = 8


* Must contain at least one digit


* Must contain at least one special character (@#$)

```python
def check_password_strength(password):

    # Check minimum length of password
    if len(password) < 8:
        return "Weak Password: Password must be at least 8 characters long."

    has_digit = False
    has_special = False
    for char in password:

        # Check if password has a digit
        if char.isdigit():
            has_digit = True

        # Check if password has a special character
        if char == "@" or char == "#" or char == "$":
            has_special = True

    if has_digit == False:
        return "Weak Password: Must contain at least one digit."

    if has_special == False:
        return "Weak Password: Must contain at least one special character (@ # $)."

    return "Strong Password "


# Taking input from user
password = input("Enter your password: ")
result = check_password_strength(password)
print(result)
```
